```python
n = int(input("Nhập n: "))
tong = 0
for i in range(2, n + 1, 2):
    tong += i
print("Tổng các số chẵn nhỏ hơn hoặc bằng n là:", tong)
```

```python
s = input()

words = s.split()

st = len(words)
tdn = max(words, key=len)

print("Số từ:", st)
print("Từ dài nhất:", tdn )
```

```python
day_so = list(map(int, input().split()))
sorted_list = sorted(set(day_so))
print("Thứ tự các số:", sorted_list)
if len(sorted_list) >= 2:
    print("Số lớn thứ hai là:", sorted_list[-2])
else:
    print("Không tồn tại số lớn thứ hai")
```

```python
ds_diem = list(map(float, input("Nhập điểm: ").split()))
diem_hop_le = [s for s in ds_diem if 0 <= s <= 10]
if not diem_hop_le:
    print("Không có điểm hợp lệ")
else:
    diem_TB = sum(diem_hop_le) / len(diem_hop_le)
    dat = sum(1 for s in diem_hop_le if s >= 5)
    print("Danh sách điểm hợp lệ:", diem_hop_le)
    print("Điểm trung bình:", round(diem_TB, 2))
    print("Số sinh viên đạt:", dat)
```
