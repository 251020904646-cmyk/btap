```python
numbers = []

print("Nhập vào 10 số nguyên:")

for i in range(10):
    num = int(input(f"Số thứ {i+1}: "))
    numbers.append(num)
so_am = sum(1 for x in numbers if x < 0)
so_duong = sum(1 for x in numbers if x > 0)
so_0= sum(1 for x in numbers if x == 0)

print("\nKết quả:")
print(f"Số âm: {so_am}")
print(f"Số dương: {so_duong}")
print(f"Số bằng 0: {so_0}")
```

```python
n = int(input("Nhập một số từ 1 đến 9: "))

if 1 <= n <= 9:
    print(f"Bảng cửu chương của {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")
else:
    print("Vui lòng nhập số trong khoảng 1 đến 9!")
```

```python
n = int(input("Nhập một số nguyên n: "))

if n < 2:
    print(f"{n} không phải là số nguyên tố.")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{n} là số nguyên tố.")
    else:
        print(f"{n} không phải là số nguyên tố.")
```
